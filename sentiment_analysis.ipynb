```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import string
import re
from textblob import Word
def rm_punctuation_stopwords(text):
    text = text.lower()
    text = re.sub(r'[\n\d]+', ' ', text)
    text = re.sub(r'[^\x20-\x7E]', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    stop_words = nltk.corpus.stopwords.words('english')
    stop_words.extend(['rated', 'ratedn'])
    text = ' '.join(word for word in text.split() if word not in stop_words)
    text = " ".join([Word(word).lemmatize() for word in text.split()])
    return text

df['reviews_list'] = df['reviews_list'].progress_apply(rm_punctuation_stopwords)
```

```python
from textblob import TextBlob
df['polarity'] = df["reviews_list"].progress_apply(lambda x: TextBlob(x).sentiment[0])
df['subjectivity'] = df["reviews_list"].progress_apply(lambda x: TextBlob(x).sentiment[1])

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
df['sentiment'] = df['reviews_list'].progress_apply(lambda x: analyzer.polarity_scores(x))
df = df.join(df['sentiment'].progress_apply(pd.Series))
```

```python
df.drop(columns = ['sentiment', 'subjectivity', 'neg', 'neu', 'pos'], inplace = True)
df.sample(3)
```

```python
df.rename(columns = {'city' : 'location'}, inplace = True)
df.dropna(inplace = True)
```

```python
df['super_score'] = df.progress_apply(lambda x: (x['polarity'] * x['compound']) + x['rate'], axis = 1)
ss_min = df['super_score'].min()
ss_max = df['super_score'].max()
df['super_score'] = df['super_score'].progress_apply(lambda x: 5 * (x - ss_min) / (ss_max- ss_min))
```

```python
df.drop_duplicates(subset = ['name', 'super_score', 'address'], inplace = True)
df.sort_values(by = 'super_score', ascending = False, inplace = True)
df.reset_index(drop = True, inplace = True)
```
